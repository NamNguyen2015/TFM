```python
import pandas as pd
import numpy as np
```

```python
# Define the file path
file_path = '../datas/Option_B/Brazil_Energy balance matrix.xlsx'

# Read all sheets into a dictionary of DataFrames
Dict = pd.read_excel(file_path, sheet_name=None, skiprows=range(4), skipfooter=3)

years=[]
# Access individual sheets by their names
for k,df in Dict.items(): 
    # append keys to a list 
    years.append(k)
    #Calculate the number of NaN values in each column
    nan_counts = df.isnull().sum()
    
    # Replace NaN values in each column with the desired value (e.g., 0)
    for column, count in nan_counts.items():
        if count > 0:
            df[column].fillna(0, inplace=True)
            
    # Round the values of each column to 2 decimal places
    df = df.applymap(lambda x: round(x, 2) if isinstance(x, (int, float)) else x)
   # df=df.applymap(lambda x: round(x, 2) if pd.notnull(x) and isinstance(x, (int, float)) else x)
    

Dict[years[0]]
  
```
