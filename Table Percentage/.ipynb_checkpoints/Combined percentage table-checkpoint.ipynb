```python
import pandas as pd
import numpy as np

# Generating a dataframe with random values between 0 and 10
df = pd.DataFrame({
    'A': np.random.randint(0, 11, 5),
    'B': np.random.randint(0, 11, 5),
    'C': np.random.randint(0, 11, 5)
})

# Adding a 'Total' column which is the sum of the rows
df['Total'] = df.sum(axis=1)

df
```

```python
# Calculating the percentage of each column based on the 'Total' of each row
df_percentage_rowwise = df[['A', 'B', 'C']].div(df['Total'], axis=0) * 100

# Adding the 'Total' column as 100% for each row, since it's the cumulative percentage
df_percentage_rowwise['Total'] = 100

df_percentage_rowwise
```

```python
# Initializing an empty DataFrame for the final combined table
final_df = pd.DataFrame()

# Iterating through each column in the original dataframe
for col in df.columns:
    # Constructing the paired data for 'Value' and 'Percentage'
    paired_data = pd.concat([df[col], df_percentage_rowwise[col]], axis=1)
    paired_data.columns = pd.MultiIndex.from_product([[col], ['Value', 'Percentage']])
    # Adding the paired data to the final dataframe
    final_df = pd.concat([final_df, paired_data.round(2)], axis=1)

final_df
```
